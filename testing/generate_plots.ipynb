```python
import matplotlib.pyplot as plt
```

```python
avg_latencies = []
match_throughput = []
avg_mmr_discrepancy = []
std_mmr_discrepancy = []
avg_churn_likelihood = []
std_churn_likelihood = []
```

```python
# Your actual x-axis: number of players joining
x_values = list(range(1000, 10001, 1000))  # [1000, 2000, ..., 10000]

# Mapping metrics to labels and colors
metrics = {
    "Average Latency (ms)": (avg_latencies, 'tab:blue'),
    "Match Throughput": (match_throughput, 'tab:orange'),
    "Avg MMR Discrepancy": (avg_mmr_discrepancy, 'tab:green'),
    "Std MMR Discrepancy": (std_mmr_discrepancy, 'tab:red'),
    "Avg Churn Likelihood": (avg_churn_likelihood, 'tab:purple'),
    "Std Churn Likelihood": (std_churn_likelihood, 'tab:brown')
}

for title, (data, color) in metrics.items():
    plt.figure()
    plt.plot(x_values, data, marker='o', color=color)
    plt.title(title)
    plt.xlabel("Number of Players Joining")
    plt.ylabel(title)
    plt.grid(True)
    filename = title.lower().replace(" ", "_").replace("(", "").replace(")", "") + ".png"
    plt.savefig(filename)
    plt.show()
```
